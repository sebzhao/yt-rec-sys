```python
def add_attr_from_pandas(G, df, id, columns):
    ids = df[id]
    inner_dicts = [{column:df[df[id] == i][column].iloc[0] for column in columns} for i in ids]
    attribute_dict = dict(zip(ids, inner_dicts))
    print(attribute_dict)
    nx.set_node_attributes(G, attribute_dict)
```

```python
def add_attr_from_pandas(G, df, id, columns):
    ids = df[id]
    vals = {column: df[column] for column in columns}
    inner_dicts = [{column:vals[column].iloc[i] for column in columns} for i in range(len(ids))]
    attribute_dict = dict(zip(ids, inner_dicts))
    print(attribute_dict)
    nx.set_node_attributes(G, attribute_dict)
```
